オンライン学習の定義




オンライン学習とは学習データが入ってくる度にパラメータ（weight）を更新し、学習を進めていく方法。確率的勾配降下法で用いることができ、日に日にデータが蓄積していったりするような場合に役立つ。一方、バッチ学習では一度にすべての学習データを使用して勾配降下法を用いてパラメータ更新を行う。そのため計算コストがかかり、データコストも掛かる。なので、深層学習ではオンライン学習が重要。
